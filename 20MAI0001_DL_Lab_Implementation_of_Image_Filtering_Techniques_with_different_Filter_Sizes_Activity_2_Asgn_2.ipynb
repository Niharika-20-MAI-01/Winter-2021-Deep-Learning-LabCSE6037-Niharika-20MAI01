To find the ENTROPY of an Image as a Metric to measure or determine the quality of the Image

```python
#Function to find the Entopy of an Image
def image_entropy(img):
  
    histogram = img.histogram()
    histogram_length = sum(histogram)
 
    samples_probability = [float(h) / histogram_length for h in histogram]
 
    return -sum([p * math.log(p, 2) for p in samples_probability if p != 0])
 

```

Entropy Calculation for the Original Input Image:

```python
image_ent = Image.open('catyp.jpg')
image_entropy(image_ent)
```

Implementation of Image Filtering Technique using (3 x 3) and (5 x 5) Image Filter sizes :

Implementation of 3x3 image filtering with Box Filter as hyperparameter :

```python
# 3x3 image filtering technique
def plot_cv_img_3(input_image,output_image):
  fig, ax = plt.subplots(nrows=1, ncols=2)
  ax[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
  ax[0].set_title('Input Image')
  ax[0].axis('off')
  ax[1].imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
  ax[1].set_title('Filtererd Image (3,3)')
  ax[1].axis('off')
  plt.savefig('op2.png')
  plt.show()
  
```

```python
#Changing Kernel size to (3 x 3)
kernel_size3 = (3,3)
# opencv implementation 
blur3 = cv2.blur(img, kernel_size3)
# plot
plot_cv_img_3(img, blur3) 
```
