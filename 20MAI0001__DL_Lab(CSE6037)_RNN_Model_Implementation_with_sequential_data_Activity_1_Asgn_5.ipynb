```python
# Reading the Text Stream data content from the uploaded written/created text file : text-5.txt
with open('/content/drive/My Drive/text-5.txt', 'r') as f: 

#To Read the data and to turn the data into lower case  
    data = f.read().lower()
    print(data)
```

#Input data formatting :

```python
#To get the set of characters used in the data and sorts them
chars = sorted(list(set(data)))

#To store the total number of characters in the entire dataset into totalChars
totalChars = len(data)

#To store the number of unique characters or the length of chars
numberOfUniqueChars = len(chars)
```

```python
#creates and stores a sorted list of characters used in the text
chars
```

#To view the Input data after application of data formatting on it :

```python
data
```

```python
#This allows for characters to be represented by numbers
CharsForids = {char:Id for Id, char in enumerate(chars)}
```

```python
#This is the opposite to the above the number is the key and the corresponding character is the value
idsForChars = {Id:char for Id, char in enumerate(chars)}
```

```python
#Basically how many characters we want one training example to contain or in other words the number of time-steps.
numberOfCharsToLearn = 100
```

```python
#creates a dictionary where each character is a key. Each key character is represented by a number
CharsForids["o"]
```

To create empty lists for storing the formatted data as input, charX and output, y

```python
#Input data
charX = []
```

```python
#Output data
y = []
```

creates a counter for our for loop. We run our loop for a 100 (numberOfCharsToLearn) less as we will be referencing the last 100 as the output chars or the consecutive chars to the input

```python
#Since our timestep sequence represetns a process for every 100 chars we omit
#the first 100 chars so the loop runs a 100 less or there will be index out of
#range
counter = totalChars - numberOfCharsToLearn
```

The InputChars stores the first 100 chars and then as the loop iterates, it takes the next 100 and so on

The OutputChars stores only 1 char, the next char after the last char in theInputChars

The charX list is appended to with 100 integers. Each of those integers are IDs of the chars in theInputChars

y.append appends an integer ID every iteration to the y list corresponding to the single char in theOutputChars

```python
#This loops through all the characters in the data skipping the first 100
for i in range(0, counter, 1):
#This one goes from 0-100 so it gets 100 values starting from 0 and stops
#just before the 100th value
  theInputChars = data[i:i+numberOfCharsToLearn]
#With no ':' you start with 0, and so you get the actual 100th value
#Essentially, the output Chars is the next char in line for those 100 chars in charX
  theOutputChars = data[i + numberOfCharsToLearn]
#Appends every 100 chars ids as a list into charX
  charX.append([CharsForids[char] for char in theInputChars])
#For every 100 values there is one y value which is the output
  y.append(CharsForids[theOutputChars])
```

Shapes the input array into [samples, time-steps, features], required for Keras

```python
#Len(charX) represents how many of those time steps we have
#The numberOfCharsToLearn is how many character we process
#Our features are set to 1 because in the output we are only predicting 1 char
X = np.reshape(charX, (len(charX), numberOfCharsToLearn, 1))
```

A form of normalisation :

```python
#This is done for normalization
X = X/float(numberOfUniqueChars)
```

This converts y into a one-hot vector. A one-hot vector is an array of 0s and 1s. The 1 only occurs at the position where the ID is true. For example, say we have 5 unique character IDs, [0, 1, 2, 3, 4].

```python
#This sets it up for us so we can have a categorical(#feature) output format
y = np_utils.to_categorical(y)
```

#Building / creating  the Recurrent Neural Network (RNN) Model

This uses the Sequential() import. This essentially initialises the network. It creates an empty "template model".

```python
model_rnn = Sequential()
```

```python
#Since we know the shape of our Data we can input the timestep and feature data
```

#To add the Layers to the RNN Model being built:

we now add our first layer to the empty "template model". This is the LSTM layer which contains 256 LSTM units, with the input shape being input_shape=(numberOfCharsToLearn, features).

This "drops-out" a neuron. The 0.2 represents a percentage, it means 20% of the neurons will be "dropped" or set to 0


```python

model_rnn.add(LSTM(256, input_shape=(X.shape[1], X.shape[2])))
model_rnn.add(Dropout(0.2))
```

#To add number of features on the output layer of the RNN Model being built :

The layer acts as an output layer. It performs the activation of the dot of the weights and the inputs plus the bias

```python

model_rnn.add(Dense(y.shape[1], activation='softmax'))
```

#To observe the built RNN Model Summary :

```python
model_rnn.summary()
```

#To compile the RNN Model  :

This is the configuration settings. Our loss function is the "categorical_crossentropy" and the optimizer is "Adam"

```python

model_rnn.compile(loss='categorical_crossentropy', optimizer='adam')
```

#TO Train the Recurrent Neural Network (RNN) Model built :

This runs the training algorithm. The epochs are the number of times we want each of our batches to be evaluated.

The batch size is the how many of our input data set we want evaluated at once. In this case we input 128 of examples into the training algorithm then the next 128 and so on

```python
#The number of timestep sequence are dealt with in the fit function
model_rnn.fit(X, y, epochs=10, batch_size=128)
```

#Generation of new text by running / implementing the our RNN Model   : 

```python
#Generates a random value from 0 to anything between the length of the input data minus 1
randomVal = np.random.randint(0, len(charX)-1)

#provides us with our starting sentence in integer form
randomStart = charX[randomVal]
```

```python
#To generate 500 chars
for i in range(500): 

#generates a single data example which we can put through to predict the next char
  x = np.reshape(randomStart, (1, len(randomStart), 1))

#To normalise the single example and then put it through the prediction model
  x = x/float(numberOfUniqueChars)
  pred = model_rnn.predict(x)

#gives us back the index of the next predicted character after that sentence  
  index = np.argmax(pred)

#appending our predicted character to our starting sentence gives us 101 chars  
  randomStart.append(index)
  randomStart = randomStart[1: len(randomStart)]

#Loops until it's reached 500 and then prints out the generated text by converting the integers back into chars  
print("".join([idsForChars[value] for value in randomStart]))
```
